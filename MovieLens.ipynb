```python
# 파일 경로
movies_url = "https://raw.githubusercontent.com/wesm/pydata-book/3rd-edition/datasets/movielens/movies.dat"
ratings_url = "https://raw.githubusercontent.com/wesm/pydata-book/3rd-edition/datasets/movielens/ratings.dat"
users_url = "https://raw.githubusercontent.com/wesm/pydata-book/3rd-edition/datasets/movielens/users.dat"

# 데이터프레임으로 불러오기
users = pd.read_table(users_url, sep='::', header=None, names=["user_id", "gender", "age", "occupation", "zip"], engine='python')
ratings = pd.read_table(ratings_url, sep='::', header=None, names=["user_id", "movie_id", "rating", "timestamp"], engine='python')
movies = pd.read_table(movies_url, sep='::', header=None, names=["movie_id", "title", "genres"], engine='python')
```

```python
users.head(5)
ratings.head(5)
movies.head(5)
ratings
```

```python
data = pd.merge(pd.merge(ratings, users), movies)
data
data.iloc[0]
```

```python
mean_ratings = data.pivot_table("rating", index="title",
                                columns="gender", aggfunc="mean")
mean_ratings.head(5)
```

```python
ratings_by_title = data.groupby("title").size()
```

```python
ratings_by_title.head()
```

```python
active_titles = ratings_by_title.index[ratings_by_title>=250]
```

```python
active_titles
```

```python
mean_ratings = mean_ratings.loc[active_titles]

mean_ratings
```

```python
top_female_ratings = mean_ratings.sort_values("F", ascending=False)
```

```python
top_female_ratings.head()
```

```python
mean_ratings["diff"] = mean_ratings["M"] - mean_ratings["F"]
```

```python
sorted_by_diff = mean_ratings.sort_values("diff")
```

```python
sorted_by_diff.head()
```

```python
sorted_by_diff[::-1].head()
```

```python
rating_std_by_title = data.groupby("title")["rating"].std()
```

```python
rating_std_by_title = rating_std_by_title.loc[active_titles]
```

```python
rating_std_by_title.head()
```

```python
rating_std_by_title.sort_values(ascending=False)[:10]
```

```python
movies["genres"].head()
```

```python
movies["genres"].head().str.split("|")
```

```python
movies["genre"] = movies.pop("genres").str.split("|")

movies.head()
```

```python
movies_exploded = movies.explode('genre')
```

```python
movies_exploded[:10]
```

```python
ratings_with_genre = pd.merge(pd.merge(movies_exploded, ratings), users)
```

```python
ratings_with_genre.iloc[0]
```

```python
genre_ratings = (ratings_with_genre.groupby(["genre", "age"])["rating"].mean().unstack("age"))
```

```python
genre_ratings[:10]
```
